```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
# central tendency --> mean , median , mode
```

```python
# measure of dispersion  --> range , std , var , coef of var

# var 
print("varience (flipper len : )",df['flipper_length_mm'].var())
print("std : ", df['flipper_length_mm'].std())
```

---
#### how to know std is small or not 
- method 1 : 
  - compare with mean 
- method 2 :
  - coeff of variance


```python
avg_flipper_len = df['flipper_length_mm'].mean()
std_of_flipper_len = df['flipper_length_mm'].std()
print("method 1 : ")
print('avg',avg_flipper_len , "std - ", std_of_flipper_len)
# data is cluster or close to each other

coef_of_var = (std_of_flipper_len/avg_flipper_len)*100
print("method 2 :")
print(f'coef of var : {coef_of_var:.2f}%')  # data is close to each other if %age is small
```
