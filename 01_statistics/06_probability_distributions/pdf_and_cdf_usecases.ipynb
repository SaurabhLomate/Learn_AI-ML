```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
# sepal length
sns.kdeplot(data = iris, x = 'sepal_length', hue='species')
plt.show()


print("inference : \nwith sepal_width it is impossible to distinguish between species of the flower because the graphs are highly overlapping with each other")
```

```python
# petal length
sns.kdeplot(data = iris, x = 'petal_length', hue='species')
plt.show()
print("inference : \nwith the help of this graph we can identify or our model can identify which kind of flower is this \nusing petal length the model can distinguish between differnt flower species like if the petal length is between 1- 2 then the species would be setosa , if the petal length is between 2 - 4.8 the flower would be of species versicolor else virginica")
```

```python
# sepal width
sns.kdeplot(data = iris, x = 'sepal_width', hue='species')
plt.show()

print("inference : \nwith sepal_width it is impossible to distinguish between species of the flower because the graphs are highly overlapping with each other")
```

```python
# petal width
sns.kdeplot(data = iris, x = 'petal_width', hue='species')
plt.show()

print("inference : \nwith the help of this graph we can identify or our model can identify which kind of flower is this \nusing petal width the model can distinguish between differnt flower species like if the petal width is between 0.0 - 0.7 then the species would be setosa , if the petal width is between 0.7 - 1.6 the flower would be of species versicolor else virginica")
```

### CDF usecases

```python
sns.kdeplot(data = iris, x = 'petal_width', hue='species')
sns.ecdfplot(data= iris, x = 'petal_width' , hue = 'species')
plt.show()
print("Inference : \nthe flower with petal width between 1.0 - 1.7 is of type vesicolor \n at point 1.7 we can say that 0.99 means (99%) chance is that it is a versicolor species flower and 1% chance of being the verginica type\nSo with this we can say that we are 99% correct or chances or sure that the flower would be of species versicolor")
```
