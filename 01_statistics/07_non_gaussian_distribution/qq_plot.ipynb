```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

manually draw qq-plot

```python
iris = sns.load_dataset("iris")
sns.kdeplot(data= iris, x='sepal_length')  # approx normally distributed
```

```python
# sort data and calculate quantiles --> out data
x = sorted(iris.sepal_length.tolist())
x_quant = []

for i in range(1,101) :
    x_quant.append(np.percentile(x,i))
```

```python
# generate normal dist data
# y = np.random.normal(loc=0 , scale=1, size=1000)
# y = np.random.randn(1000)
y = np.random.standard_normal(1000)
```

```python
y_quant = []
for i in range(1,101):
    y_quant.append(np.percentile(y,i))
```

```python
# plot graph

sns.scatterplot(x= x_quant, y= y_quant)
```

using statsmodel 

```python
fig = stats.probplot(x= x, dist='norm',fit=True,plot=plt)
```

```python
# 2nd method
import statsmodels.api as sm
fig = sm.qqplot(data= iris['sepal_length'], line='45', fit=True)
```
