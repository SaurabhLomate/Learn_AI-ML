````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
````

standard normal distribution 

Let me explain each part of the code for plotting a normal distribution with its curve:
``` python
    # Generate the normal distribution data
    x = np.random.normal(loc=50, scale=10, size=1000)
```
loc=50: Sets the mean (center) of the distribution to 50.  
scale=10: Sets the standard deviation (spread) to 10.  
size=1000: Generates 1000 random numbers.  

```python
    plt.hist(x, bins=30, density=True, alpha=0.7, color='skyblue', label='Histogram')
```
bins=30: Divides data into 30 equal-width bars.  
density=True: Normalizes the histogram so area equals 1.  
alpha=0.7: Sets transparency (0=transparent, 1=solid).  
color='skyblue': Sets histogram color.  
label='Histogram': Adds label for legend.  

```python
    xmin, xmax = plt.xlim()
    x_curve = np.linspace(xmin, xmax, 100)
```
Gets the x-axis range from the histogram.  
Creates 100 evenly spaced points for smooth curve.  

```python
    plt.plot(x_curve, stats.norm.pdf(x_curve, loc=50, scale=10), 
         'r-', lw=2, label='Normal Distribution')
```

stats.norm.pdf(): Calculates probability density function.  
loc=50, scale=10: Same parameters as random data.  
'r-': Red solid line.  
lw=2: Line width of 2.  
label='Normal Distribution': Adds label for legend.  

````python
# Generate the normal distribution data
x = np.random.normal(loc=50, scale=10, size=1000)

# Create the histogram with density=True to normalize the data
plt.hist(x, bins=30, density=True, alpha=0.7, color='skyblue', label='Histogram')

# Create x axis points for the curve
xmin, xmax = plt.xlim()
x_curve = np.linspace(xmin, xmax, 100)

# Add the normal distribution curve
plt.plot(x_curve, stats.norm.pdf(x_curve, loc=50, scale=10), 
         'r-', lw=2, label='Normal Distribution')

# Customize the plot
plt.grid(True)
plt.legend()
plt.title('Normal Distribution with Curve')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
````

````python
# Method 1: Using numpy's built-in functions
x = np.random.normal(loc=50, scale=10, size=1000)
plt.hist(x, bins=30, density=True, alpha=0.7, color='skyblue')

# Generate points for smooth curve using numpy
x_points = np.linspace(x.min(), x.max(), 100)
y_points = 1/(10 * np.sqrt(2 * np.pi)) * np.exp(-(x_points - 50)**2 / (2 * 10**2))
plt.plot(x_points, y_points, 'r-', lw=2)
plt.show()

# Method 2: Using seaborn (easier and more elegant)
import seaborn as sns
sns.histplot(x, stat='density', kde=True)
plt.show()

# Method 3: Using seaborn's distplot (combines histogram and KDE)
sns.kdeplot(x, color='red')
plt.hist(x, bins=30, density=True, alpha=0.7)
plt.show()
````

````python
data_mean=50
data_std=10
data_size=500
np.random.seed(0)
data=np.random.normal(loc=data_mean,scale=data_std,size=data_size)

plt.figure(figsize=(5,5))
plt.hist(data,bins=30,density=True,color='pink',alpha=0.6)
plt.legend('hist')
plot_data=np.linspace(data.min(),data.max(),100)

# #probability density function
pdf= (1 / (data_std * np.sqrt(2 * np.pi))) * np.exp( - (plot_data - data_mean)**2 / (2 * (data_std)**2))

plt.plot(plot_data , pdf , 'b-')
plt.legend(['normal_distribution','data'])
plt.show()
````

**probability distribution**

probability mass function - for descrete variables

````python
# lets take an example of an dice
df=pd.DataFrame(np.random.randint(1,7,1000),columns=['possible_outcomes'])
probability=pd.DataFrame(df.groupby(by=['possible_outcomes']).value_counts().rename("probabilities")/df.size).reset_index()
plt.figure(figsize=(6,3))
plt.bar(probability['possible_outcomes'],probability['probabilities'])
plt.xlabel('possible_outcomes',color='red')
plt.ylabel('probabilities',color='red')
plt.title('Probability mass Function (PMF)',color='blue')

plt.show()
````

Cumulative distribution function -- for descrete data
___ 

````python
probability['cumulative_probability']=np.cumsum(probability['probabilities'])
probability
````

````python
plt.figure(figsize=(6,3))
plt.bar(probability['possible_outcomes'],probability['cumulative_probability'])
plt.xlabel('possible outcomes' ,color='red')
plt.ylabel('cumulative probabilities', color='red')
plt.title('Cumulative Distribution function (CMF)', color='green')
plt.tight_layout()
plt.show()
````
